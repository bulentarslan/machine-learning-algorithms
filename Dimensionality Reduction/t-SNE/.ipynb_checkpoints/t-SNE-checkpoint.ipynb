# t-SNE vs PCA on Wine dataset

```python
# t-SNE

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

# Importing the dataset
dataset = pd.read_csv('Wine.csv')
X = dataset.iloc[:, 0:13].values
y = dataset.iloc[:, 13].values

# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Applying t-SNE
from sklearn.manifold import TSNE
tsne = TSNE(n_components = 2)
X_train_tsne = tsne.fit_transform(X_train)
#X_test = tsne.transform(X_test) # tsne has no attribute 'transform'
kl_divergence = tsne.kl_divergence_
```

```python
# Applying PCA
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X_train_pc = pca.fit_transform(X_train)
X_test_pc = pca.transform(X_test)
explained_variance = pca.explained_variance_ratio_
```

```python
# Visualizing t-SNE vs PCA
# Visualising t-SNE for Training set results 
from matplotlib.colors import ListedColormap
f = plt.figure(figsize=(15,6))
f.add_subplot(1,2,1)
X_set, y_set = X_train_tsne, y_train
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
plt.title('Projection obtained by t-SNE')
plt.xlabel('Comp1')
plt.ylabel('Comp2')
plt.legend()

# Visualizing PCs
# Visualising PCA for Training set results
from matplotlib.colors import ListedColormap
f.add_subplot(1,2,2)
X_set, y_set = X_train_pc, y_train
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
plt.title('Projection obtained by PCA')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.show()
```

```python
#Visualizing t-SNE with different parameters
f = plt.figure(figsize=(10,20))
h = 1
plt.subplots_adjust(hspace=.5)
for perp in np.random.random(10,)*50: #unfiormly sample values for perplexity between 0-50
    tsne = TSNE(perplexity=perp,random_state=1)
    X_train_t=tsne.fit_transform(X_train)
    from matplotlib.colors import ListedColormap
    X_set, y_set = X_train_t, y_train
    f.add_subplot(5,2,h)
    for i,j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
    plt.title('t-SNE perplexity={:10.2f}'.format(perp))
    plt.xlabel('Comp1')
    plt.ylabel('Comp2')
    plt.legend()
    h +=1
    
plt.show()
```

```python
#Visualizing t-SNE with different parameters
f = plt.figure(figsize=(10,20))
h = 1
plt.subplots_adjust(hspace=.5)
for alpha in np.random.random(10,)*200: #unfiormly sample values for perplexity between 0-50
    tsne = TSNE(learning_rate=alpha, random_state=1)
    X_train_t=tsne.fit_transform(X_train)
    from matplotlib.colors import ListedColormap
    X_set, y_set = X_train_t, y_train
    f.add_subplot(5,2,h)
    for i,j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
    plt.title('t-SNE learning_rate={:10.2f}'.format(alpha))
    plt.xlabel('Comp1')
    plt.ylabel('Comp2')
    plt.legend()
    h +=1
    
plt.show()
```

```python
#Visualizing t-SNE with different methods
f = plt.figure(figsize=(15,4))
h = 1
plt.subplots_adjust(hspace=.5)
for met in ['barnes_hut','exact']: #unfiormly sample values for perplexity between 0-50
    tsne = TSNE(method=met,random_state=1)
    X_train_t=tsne.fit_transform(X_train)
    from matplotlib.colors import ListedColormap
    X_set, y_set = X_train_t, y_train
    f.add_subplot(1,2,h)
    for i,j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
    plt.title('t-SNE method={}'.format(met))
    plt.xlabel('Comp1')
    plt.ylabel('Comp2')
    plt.legend()
    h +=1
    
plt.show()
```

```python
#Visualizing t-SNE with different random initializations
f = plt.figure(figsize=(10,20))
h = 1
plt.subplots_adjust(hspace=.5)
for ran in np.random.randint(10,high=60,size=(10,)): #unfiormly sample values for seed
    tsne = TSNE(random_state=ran)
    X_train_t=tsne.fit_transform(X_train)
    from matplotlib.colors import ListedColormap
    X_set, y_set = X_train_t, y_train
    f.add_subplot(5,2,h)
    for i,j in enumerate(np.unique(y_set)):
        plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1],
                c = ListedColormap(('red', 'green', 'blue'))(i), label = j)
    plt.title('t-SNE seed={}'.format(ran))
    plt.xlabel('Comp1')
    plt.ylabel('Comp2')
    plt.legend()
    h +=1
    
plt.show()
```

## Conclusion: 
- Perplexity: Expected cluster density. Apart from extemely low values, t-SNE is very much insensitive to this parameter.
- Learning Rate: Increase to prevent model from falling into local minimas. t-SNE algorithm did a very good job on various values ranging from 0 to 200. No drawbacks observed for any particular range.
- Method: No significant accuracy issue. Barnes-Hut method did slightly better.
- Random Initialization: Has significant effect on clusters' placement across projection space. Trying out a several seeds would help a better visualization.
